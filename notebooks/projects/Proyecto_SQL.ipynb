```python
import pandas as pd
import os
from pandasql import sqldf
import sqlite3
```

En este proyecto simularemos de una forma resumida como una tienda llamada Interpaq lleva el registro de de sus clientes en una tabla llamada "clientes" y las ventas dadas en una fecha, esta tabla se llamará "ventas".

El objetívo es observar como las dos tablas pueden unirse en un data frame para luego hacer cualquier tipo de análisis sobre las ventas de dicha tienda.


```python
con = sqlite3.connect("interpaq.db") #Creamos el DataBase de forma local
```

```python
#Tabla para cliente
con.execute("""CREATE TABLE clientes(
  id INT PRIMARY KEY NOT NULL,
  nombre TEXT NOT NULL,
  email TEXT NOT NULL
)
""")
#Tabla datos de ventas por cliente
con.execute("""CREATE TABLE ventas(
  venta_id INT PRIMARY KEY NOT NULL,
  cliente_id INT NOT NULL,
  monto REAL NOT NULL,
  fecha TEXT NOT NULL
)""")
```
