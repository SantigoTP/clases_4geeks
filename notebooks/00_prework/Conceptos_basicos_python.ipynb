***Metodos cadena de caracteres***


```python
# ## 2. Métodos de Cadenas de Caracteres (Strings)

# Las cadenas de caracteres son inmutables en Python. Los métodos que veremos devuelven una nueva cadena con la modificación.

# ### Asignación y Concatenación
cadena1 = "Hola"
cadena2 = "Mundo"
saludo = cadena1 + " " + cadena2
print(f"Asignación y concatenación: {saludo}")

# ### Búsqueda
# `find()`: Devuelve el índice de la primera ocurrencia o -1 si no se encuentra.
# `index()`: Similar a `find()`, pero lanza un ValueError si no se encuentra.
# `count()`: Cuenta las ocurrencias de una subcadena.
texto = "Python es un lenguaje de programación muy popular. Python es versátil."
print(f"\nOriginal: '{texto}'")
print(f"Búsqueda 'lenguaje' (find): {texto.find('lenguaje')}")
print(f"Búsqueda 'Java' (find): {texto.find('Java')}")
print(f"Búsqueda 'Python' (count): {texto.count('Python')}")

# ### Extracción (Substrings)
# Se realiza mediante slicing `[inicio:fin:paso]`.
cadena_original = "Programacion"
print(f"\nOriginal: '{cadena_original}'")
print(f"Extracción (índices 0 a 4): {cadena_original[0:5]}") # 'Progr'
print(f"Extracción (desde índice 6 hasta el final): {cadena_original[6:]}") # 'macion'
print(f"Extracción (últimos 3 caracteres): {cadena_original[-3:]}") # 'ion'

# ### Comparación
# Se usan operadores relacionales.
cadena_a = "manzana"
cadena_b = "banana"
cadena_c = "Manzana" # Diferente por mayúscula
print(f"\nComparación 'manzana' == 'manzana': {cadena_a == 'manzana'}")
print(f"Comparación 'manzana' == 'banana': {cadena_a == cadena_b}")
print(f"Comparación 'manzana' == 'Manzana': {cadena_a == cadena_c}") # Falso por diferencia de mayúsculas/minúsculas

# ### Otros Métodos de Cadenas
# `len()`: Longitud de la cadena.
# `strip()`, `lstrip()`, `rstrip()`: Eliminar espacios en blanco (o caracteres específicos) al inicio/final.
# `title()`, `istitle()`: Capitaliza la primera letra de cada palabra / verifica si está en formato título.
# `lower()`, `islower()`: Convierte a minúsculas / verifica si está en minúsculas.
# `upper()`, `isupper()`: Convierte a mayúsculas / verifica si está en mayúsculas.
# `swapcase()`: Invierte mayúsculas y minúsculas.
# `capitalize()`: Capitaliza solo la primera letra de la cadena.
# `center()`, `ljust()`, `rjust()`: Justificación de texto con relleno.
# `startswith()`, `endswith()`: Verifica si la cadena comienza/termina con una subcadena.

mi_texto = "   Hola Mundo Python   "
print(f"\nOriginal: '{mi_texto}' (longitud: {len(mi_texto)})")
print(f"strip(): '{mi_texto.strip()}' (longitud: {len(mi_texto.strip())})")
print(f"lstrip(): '{mi_texto.lstrip()}'")
print(f"rstrip(): '{mi_texto.rstrip()}'")

texto_capital = "este es un titulo"
print(f"\nOriginal: '{texto_capital}'")
print(f"capitalize(): '{texto_capital.capitalize()}'")
print(f"title(): '{texto_capital.title()}'")
print(f"istitle(): {texto_capital.title().istitle()}")

texto_mayus_min = "PYTHON es grandioso"
print(f"\nOriginal: '{texto_mayus_min}'")
print(f"lower(): '{texto_mayus_min.lower()}'")
print(f"islower(): {texto_mayus_min.lower().islower()}")
print(f"upper(): '{texto_mayus_min.upper()}'")
print(f"isupper(): {texto_mayus_min.upper().isupper()}")
print(f"swapcase(): '{texto_mayus_min.swapcase()}'")

cadena_justificar = "Python"
print(f"\nJustificación:")
print(f"center(20, '-'): '{cadena_justificar.center(20, '-')}'")
print(f"ljust(20, '*'): '{cadena_justificar.ljust(20, '*')}'")
print(f"rjust(20, '#'): '{cadena_justificar.rjust(20, '#')}'")

frase = "El perro corre y el gato juega"
print(f"\nstartsWith('El'): {frase.startswith('El')}")
print(f"endsWith('juega'): {frase.endswith('juega')}")
print(f"count('e'): {frase.count('e')}")
```

***Sentencias condicionales if, elif, else***


```python
# Permiten ejecutar bloques de código basados en condiciones.

# ### Ejemplo Básico
temperatura = 25

if temperatura > 30:
    print("Hace mucho calor.")
elif temperatura > 20:
    print("La temperatura es agradable.")
else:
    print("Hace frío.")

# ### Anidamiento
edad = 18
tiene_licencia = True

if edad >= 18:
    print("Eres mayor de edad.")
    if tiene_licencia:
        print("Puedes conducir.")
    else:
        print("Necesitas una licencia para conducir.")
else:
    print("Eres menor de edad.")

# ### Parámetros `end` y `sep` en `print()`
# `end`: Especifica qué imprimir al final de la salida. Por defecto es `\n` (salto de línea).
# `sep`: Especifica cómo separar los argumentos en `print()`. Por defecto es un espacio.

print("\n")
print("Hola", "mundo", end="!!!\n")
print("Este", "es", "un", "ejemplo", sep="-")
print("Uno", "Dos", "Tres", sep="...", end=".\n")
print("Continúa en la misma línea.")
```

***Bucles (While and for)***

```python
contador = 0
print("\nBucle while:")
while contador < 5:
    print(f"Contador: {contador}")
    contador += 1 # Es crucial actualizar la condición para evitar bucles infinitos

# ### Bucle `for`
# Itera sobre una secuencia (listas, tuplas, cadenas, rangos, etc.).

print("\nBucle for (en lista):")
frutas = ["manzana", "banana", "cereza"]
for fruta in frutas:
    print(f"Me gusta la {fruta}.")

print("\nBucle for (con range):")
for i in range(3): # range(3) genera 0, 1, 2
    print(f"Iteración: {i}")

# ### `break` y `continue`

# `break`: Termina el bucle completamente.
print("\nEjemplo de 'break' en bucle for:")
for num in range(10):
    if num == 5:
        print("Se encontró el 5. Terminando el bucle.")
        break
    print(num)

# `continue`: Salta la iteración actual y pasa a la siguiente.
print("\nEjemplo de 'continue' en bucle for:")
for num in range(5):
    if num == 2:
        print("Saltando el 2.")
        continue
    print(num)
```

***Estructuras de datos: Listas (Arrays)***


```python
# Las listas son colecciones ordenadas y mutables de elementos.

# ### Listas 1D (unidimensionales)
mi_lista = [10, 20, 30, 40, 50]
print(f"\nLista original: {mi_lista}")

# ### Acceso a elementos (índices)
print(f"Primer elemento: {mi_lista[0]}")
print(f"Último elemento: {mi_lista[-1]}")

# ### Arrays 2D (Listas de Listas)
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(f"\nMatriz 2D: {matriz}")
print(f"Elemento [0][0]: {matriz[0][0]}") # 1
print(f"Elemento [1][2]: {matriz[1][2]}") # 6

# ### Agregar elementos
# `append()`: Agrega al final.
mi_lista.append(60)
print(f"\nDespués de append(60): {mi_lista}")

# `insert(indice, valor)`: Inserta en una posición específica.
mi_lista.insert(1, 15)
print(f"Después de insert(1, 15): {mi_lista}")

# ### Eliminar elementos
# `pop()`: Elimina y devuelve el último elemento.
elemento_eliminado = mi_lista.pop()
print(f"\nDespués de pop() (elemento eliminado: {elemento_eliminado}): {mi_lista}")

# `pop(indice)`: Elimina y devuelve el elemento en un índice específico.
elemento_eliminado_indice = mi_lista.pop(1)
print(f"Después de pop(1) (elemento eliminado: {elemento_eliminado_indice}): {mi_lista}")

# `remove(valor)`: Elimina la primera ocurrencia de un valor.
mi_lista.remove(40)
print(f"Después de remove(40): {mi_lista}")

# `del`: Elimina por índice o slice.
del mi_lista[0]
print(f"Después de del mi_lista[0]: {mi_lista}")

# ### Invertir lista
mi_lista.reverse()
print(f"\nLista invertida: {mi_lista}")

# ### Ordenar elementos
mi_lista_desordenada = [5, 2, 8, 1, 9]
mi_lista_desordenada.sort() # Ordena en su lugar
print(f"\nLista ordenada (sort()): {mi_lista_desordenada}")

otra_lista_para_ordenar = [100, 50, 200]
lista_ordenada_nueva = sorted(otra_lista_para_ordenar) # sorted() devuelve una nueva lista ordenada
print(f"Lista original para sorted(): {otra_lista_para_ordenar}")
print(f"Nueva lista ordenada (sorted()): {lista_ordenada_nueva}")

# ### Búsqueda de elementos
# `in`: Para verificar si un elemento existe.
# `index()`: Para encontrar el índice de un elemento.
print(f"\n¿Existe 30 en la lista? {30 in mi_lista}")
print(f"Índice de 50: {mi_lista.index(50)}")

# ### Concatenar listas
lista1 = [1, 2, 3]
lista2 = [4, 5, 6]
lista_concatenada = lista1 + lista2
print(f"\nListas concatenadas: {lista_concatenada}")
```

***Diccionarios python***


```python
# ### Creación y Acceso
persona = {
    "nombre": "Juan",
    "edad": 30,
    "ciudad": "Madrid"
}
print(f"\nDiccionario: {persona}")
print(f"Nombre: {persona['nombre']}")
print(f"Edad: {persona['edad']}")

# ### `items()`, `keys()`, `values()`
# `items()`: Devuelve una vista de los pares clave-valor.
# `keys()`: Devuelve una vista de las claves.
# `values()`: Devuelve una vista de los valores.

print(f"\nItems: {persona.items()}")
print(f"Keys: {persona.keys()}")
print(f"Values: {persona.values()}")

# ### Recorrer un diccionario
print("\nRecorriendo diccionario:")
for key, value in persona.items():
    print(f"{key}: {value}")

print("\nRecorriendo solo claves:")
for key in persona.keys():
    print(key)

print("\nRecorriendo solo valores:")
for value in persona.values():
    print(value)

# ### `clear()`: Elimina todos los elementos.
dicc_temporal = {"a": 1, "b": 2}
dicc_temporal.clear()
print(f"\nDespués de clear(): {dicc_temporal}")

# ### `copy()`: Crea una copia superficial.
dicc_original = {"fruta": "manzana"}
dicc_copia = dicc_original.copy()
dicc_copia["fruta"] = "banana"
print(f"\nOriginal después de copia y modificación: {dicc_original}")
print(f"Copia: {dicc_copia}")

# ### `fromkeys(iterable, valor)`: Crea un diccionario con claves de un iterable y un valor por defecto.
claves = ["nombre", "edad", "ciudad"]
nuevo_dicc = dict.fromkeys(claves, "Desconocido")
print(f"\nfromkeys(): {nuevo_dicc}")

# ### Constructor de diccionarios `dict()`
# Se puede usar con pares clave=valor o con una lista de tuplas.
dicc_constructor1 = dict(nombre="Ana", edad=25)
print(f"\nConstructor dict() (kwargs): {dicc_constructor1}")

dicc_constructor2 = dict([('a', 1), ('b', 2)])
print(f"Constructor dict() (lista de tuplas): {dicc_constructor2}")

# ### `get(clave, valor_por_defecto)`: Acceso seguro a valores.
# No lanza error si la clave no existe, devuelve None o el valor por defecto.
print(f"\nget('nombre'): {persona.get('nombre')}")
print(f"get('pais', 'No especificado'): {persona.get('pais', 'No especificado')}") # La clave 'pais' no existe

# ### `popitem()`: Elimina y devuelve el último par (clave, valor) insertado (o arbitrario en versiones antiguas).
item_removido = persona.popitem()
print(f"\npopitem() removido: {item_removido}")
print(f"Diccionario después de popitem(): {persona}")

# ### `pop(clave, valor_por_defecto)`: Elimina y devuelve el valor de una clave específica.
edad_removida = persona.pop("edad")
print(f"\npop('edad') removido: {edad_removida}")
print(f"Diccionario después de pop('edad'): {persona}")

# ### `setdefault(clave, valor_por_defecto)`: Inserta una clave con un valor por defecto si la clave no existe.
# Si la clave ya existe, devuelve su valor y no la modifica.
cliente = {"id": 1, "nombre": "Carlos"}
email = cliente.setdefault("email", "sin_email@ejemplo.com")
print(f"\nsetdefault() (email no existía): {cliente}, email: {email}")

telefono = cliente.setdefault("telefono", "555-1234") # No se añade porque 'telefono' no existe aún
print(f"setdefault() (teléfono no existía): {cliente}, telefono: {telefono}")

# ### `update(otro_diccionario)`: Fusiona diccionarios.
dicc1 = {"a": 1, "b": 2}
dicc2 = {"c": 3, "a": 4} # 'a' se sobrescribe
dicc1.update(dicc2)
print(f"\nupdate(): {dicc1}")
```

***Tuplas en python***


```python
# Las tuplas son colecciones ordenadas e inmutables de elementos.

# ### Creación de Tuplas
mi_tupla = (1, 2, 3, "cuatro", 5.0)
print(f"\nMi tupla: {mi_tupla}")

# ### Tipo de Tupla (`type()`)
print(f"Tipo de mi_tupla: {type(mi_tupla)}")

# ### Tupla de un solo elemento (¡nota la coma!)
tupla_un_elemento = ("elemento",)
print(f"Tupla de un elemento: {tupla_un_elemento} (Tipo: {type(tupla_un_elemento)})")
# Sin la coma, sería solo una cadena:
no_es_tupla = ("elemento")
print(f"Sin la coma: {no_es_tupla} (Tipo: {type(no_es_tupla)})")


# ### Desempaquetado de Tuplas
# Asignar los elementos de una tupla a variables individuales.
coordenadas = (10, 20)
x, y = coordenadas
print(f"\nDesempaquetado: x={x}, y={y}")

# ### Segmentación de Tuplas (Slicing)
# Funciona igual que en listas y cadenas.
tupla_numeros = (10, 20, 30, 40, 50, 60)
print(f"\nOriginal: {tupla_numeros}")
print(f"Segmento (índices 1 a 3): {tupla_numeros[1:4]}") # (20, 30, 40)
print(f"Desde el inicio hasta índice 2: {tupla_numeros[:3]}") # (10, 20, 30)

# ### Recorrer Tuplas
print("\nRecorriendo tupla:")
for elemento in mi_tupla:
    print(elemento)

# ### Método `tuple()` (Constructor)
# Convierte otros iterables a tuplas.
lista_a_tupla = [1, 2, 3]
tupla_convertida = tuple(lista_a_tupla)
print(f"\nLista convertida a tupla: {tupla_convertida}")

# ### Concatenación de Tuplas
tupla_a = (1, 2)
tupla_b = (3, 4)
tupla_concatenada = tupla_a + tupla_b
print(f"\nTuplas concatenadas: {tupla_concatenada}")

# ### El método `zip()`
# Combina elementos de múltiples iterables en tuplas.
nombres = ("Ana", "Luis", "Marta")
edades = (25, 30, 22)
personas_zip = zip(nombres, edades)

print("\nUsando zip():")
for nombre, edad in personas_zip:
    print(f"{nombre} tiene {edad} años.")

# También puedes convertir el objeto zip en una lista de tuplas:
nombres2 = ("Pedro", "Maria")
edades2 = (40, 35)
lista_de_pares = list(zip(nombres2, edades2))
print(f"zip() convertido a lista de tuplas: {lista_de_pares}")
```
